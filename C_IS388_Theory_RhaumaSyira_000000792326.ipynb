### Use IQR (Interquartile Range)to find outlier in dataset and describe the result.

```python
vicage = homicide[['victim_age']]
vicage.shape
```

```python
# Menggunakan visualisasi 'Boxplot' untuk melihat apakah terdapat outlier atau tidak.
plt.figure(figsize=(10,10))
homicide.boxplot(column=['victim_age'])
plt.title('Boxplot : Outlier pada Victim_age')
plt.show()
```

```python
# Mencari quartile

Q1 = homicide['victim_age'].quantile(0.25)
Q3 = homicide['victim_age'].quantile(0.75)

Q1, Q3
```

```python
# Mencari IQR 
IQR = Q3 - Q1
IQR
```

```python
# lower_limit dan upper_limit

lowwer_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR

lowwer_limit,upper_limit
```

```python
# Menampilkan data Outlier
homicide[(homicide['victim_age'] < lowwer_limit) | (homicide['victim_age'] > upper_limit) ]
```

```python
# menampilkan tanpa outlier
homicide[(homicide['victim_age'] > lowwer_limit) & (homicide['victim_age'] < upper_limit) ]
```

```python
# Remove Outlier
# Menyimpan data no outlier ke var baru
no_outlier = homicide[(homicide['victim_age'] > lowwer_limit) & (homicide['victim_age'] < upper_limit)]
no_outlier
```

```python
# Outlier setelah di hapus
Q1 = no_outlier['victim_age'].quantile(0.25)
Q3 = no_outlier['victim_age'].quantile(0.75)

Q1, Q3
```

```python
# Mencari IQR 
IQR = Q3 - Q1
IQR
```

```python
# lower_limit dan upper_limit

lowwer_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR

lowwer_limit,upper_limit
```

```python
plt.figure(figsize=(10,10))
no_outlier.boxplot(column=['victim_age'])
plt.title('Boxplot : Outlier yang sudah dihapus')
plt.show()
```

```python
homicide.shape[0] - no_outlier.shape[0]
```

### c. Data Preprocessing and Preparation

#### a. use binning or encoding 

```python
# binning
# Membuat bins untuk mengelompokkan data umur ke dalam interval.
bin = np.linspace(min(homicide['victim_age']), max(homicide['victim_age']), 6)
bin
```

```python
# Membuat kategori Umur berdasarkan column 'victim_age'
homicide['age_category'] = np.where(
    (homicide['victim_age'] >= 16) & (homicide['victim_age'] <= 23), 'Remaja Dewasa',
    np.where((homicide['victim_age'] >= 24) & (homicide['victim_age'] <= 35), 'Dewasa Awal',
             np.where((homicide['victim_age'] >= 36) & (homicide['victim_age'] <= 59), 'Dewasa Menengah',
                      np.where(homicide['victim_age'] > 60, 'Lansia', 'Anak-Anak')))
)

homicide

# Kategori: 
# Anak-anak : 0-15 Tahun
# Remaja Dewasa : 16 - 23 Tahun
# Dewasa Awal : 24 - 35 Tahun
# Dewasa Menengah : 36 - 59 Tahun
# Lansia : 60 > Tahun
```

```python
# Data Visualization : Pie Chart

age_category_counts = homicide['age_category'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(age_category_counts, labels=age_category_counts.index, autopct='%1.1f%%')
plt.title('Distribusi Korban Pembunuhan Berdasarkan Kategori Usia')
plt.show()
```

#### Penjelasan: 
Hasil: 
- Dewasa Awal : 33.61%. 
- Remaja Dewas : 26.1%. 
- Dewasa Menengah : 25.76%. 
- Anak-Anak : 10.5%
- Lansia : 4.1%. 

Kesimpulan: Seperti pada Pie Chart diatas menunjukan bahwa korban pembunuhan berdasarkan kategori usia didominasi oleh katoegori Dewasa Awal (24 - 35 tahun)



#### b. Normalize the Data

```python
min_max = homicide[['victim_age', 'lat', 'lon']]
```

```python
# untuk normalisasi menggunakan Min Max pada kolom 'victim_age','lat', dan 'lon'

for column in min_max.columns:
    min_max[column] = (min_max[column] - min_max[column].min()) / (min_max[column].max() - min_max[column].min())
min_max.head()
```

```python
#menampilkan plot normalisasi
min_max.hist(figsize=(10, 10))
plt.title('Histogram Normalisasi Data')
plt.show()
```
